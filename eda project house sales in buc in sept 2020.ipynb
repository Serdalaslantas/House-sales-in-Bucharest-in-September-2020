```python
x = house_rentals[['rooms_count', 'useful_surface','construction_year', 'max_level', 'kitchens_count',
       'bathrooms_count', 'garages_count', 'parking_lots_count',
       'balconies_count']]
```

```python
y = house_rentals['price']
```

```python
y = pd.to_numeric(y)
```

```python
x[['rooms_count', 'useful_surface','construction_year', 'max_level', 'kitchens_count',
       'bathrooms_count', 'garages_count', 'parking_lots_count',
       'balconies_count']] = x[['rooms_count', 'useful_surface','construction_year','max_level', 'kitchens_count',
       'bathrooms_count', 'garages_count', 'parking_lots_count',
       'balconies_count']].apply(pd.to_numeric)
```

```python
x = x.fillna(0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
coeff_df = pd.DataFrame(lm.coef_,x.columns,columns=['coefficient'])
```

```python
coeff_df
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test,predictions),bins=50)
```
